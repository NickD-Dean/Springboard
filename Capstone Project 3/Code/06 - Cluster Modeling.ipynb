### Baseline model using an agglomerative clustering algorithm

Before I begin to iterate using KMeans, I want to get an understanding of whether or note an agglomarative model will perform better than KMeans. 

I should note that using the default for Agglomerative Clustering is to converge on two clusters, and to use 'ward' linkage to reducethe variance in the clusters uncovered. 

```python
model = AgglomerativeClustering()
labels = model.fit_predict(data)
```

```python
score = silhouette_score(data, labels)
print('A baseline agglomerative model provides a mean silhouette score of: ', score)
```

While an agglomerative model does not uncover serious structure in the data, it is a significant improvement over the baseline score from a KMeans algorithm of 0.0704.  

This indicates that while an agglomerative model may be better out of the box, both algorithms OR the data need to be further worked with in order to provide substantial structure with the clusters that can be used for customer segmentation. 

###  Finding the ideal 'k' for K-Means algorithm. 

The next step is to apply the elbow method in order to identify the ideal value for K with the KMeans algorithm. In order to do this I'll use the model demonstrated in a previous customer segmentation excercise from the Springboard course which can be found [here](https://github.com/NickD-Dean/Springboard/blob/bc9d26edbccb4a4fafafee14738975234cbb2b7e/Case%20Studies/Customer%20Segmentation%20&%20Clustering/Clustering%20Case%20Study%20-%20Customer%20Segmentation%20with%20K-Means%20-%20Tier%203.ipynb)

```python
# Finding the sum of squares for each value of K from 2 to 11
SS=[]
Range = list(range(2,11))
X_cols = data.to_numpy()
K_models ={}

for k in Range:
    model = KMeans(n_clusters=k)
    fit_model = model.fit_predict(X_cols)
    centroids = model.cluster_centers_
    SS.append(np.sum((X_cols - centroids[fit_model])**2))
    K_models[str(k)] = fit_model

plt.plot(Range, SS)
plt.xlabel('K')
plt.ylabel('SS Values');
```

Okay, here we can see that while there is somewhat of an elbow around k = 5, the average sum of squares is huge.  This indicates that there are clearly some other issues going on with the data set.  

I'm going to apply a similar iterative approach using the silhouette scores to attempt to visualize the best silhouette score. 

```python
Range = list(range(2,11))
avg_ss =[]
for k in Range:
    fig, ax1 = plt.subplots(1,1)
    fig.set_size_inches(9,5)
    # instatiate the model for 'k' and find labels
    model = KMeans(n_clusters=k, random_state=42)
    labels = model.fit_predict(X_cols)
    # Calculate & append avg score to list
    s_avg = silhouette_score(X_cols, labels)
    avg_ss.append(s_avg)
    # Calculate individual S-scores to plot
    sample_scores = silhouette_samples(X_cols, labels)
    # Setting axis limits for each plot
    ax1.set_xlim([-0.25, 1])
    ax1.set_ylim([0, len(X_cols) + (k + 1) * 10])
    # declare value of y for use in for loop
    y_lower=10
    for i in range(k):
        # aggregate S-scores for each cluster
        ith_cluster_vals = sample_scores[labels == i]  
        ith_cluster_vals.sort()
        size_cluster_i = ith_cluster_vals.shape[0]
        y_upper = size_cluster_i + y_lower
        # Set up color map for each plot
        cmap = cm.get_cmap("Spectral")
        color = cmap(float(i)/k)
        ax1.fill_betweenx(np.arange(y_lower, y_upper), 0, ith_cluster_vals, facecolor=color, 
                          edgecolor=color, alpha=0.7)
        # Labelling each individual cluster plot
        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10
    # Labelling each plot for each value of k
    ax1.set_title('Silhouette plots K = %d' % k)
    ax1.set_ylabel('Clusters')
    ax1.set_xlabel('Silhouette Values')
    
    ax1.axvline(x=s_avg, color='red', linestyle='--')
    ax1.set_yticks([])
    ax1.set_xticks([-0.2, 0, 0.2, 0.4, 0.6, 0.8, 1])
    plt.suptitle(("Silhouette analysis for KMeans clustering on sample data with n_clusters = %d" % k), 
                 fontsize=14, fontweight='bold')
```

Using the silhouette method to varify the results for KMeans from applying the elbow method and almost no clusters have a silhouette score that is greater than 0.2, indicating that no real structure has been found in the data.  There needs to be another approach taken here from a data standpoint. 

### Applying a log transform to the skewed numerical data

The next step towards finding valid clusters upon which to base customer segmentation is to log transform the skewed data. 

Using the pd.skew() method I'll evaluate columns which have an abnormal skewness.  The function returns a coefficient of skew for a given column, by which I'll decide which columns to apply a log transform to.   

Examples of distributions with skewness include the following. [source](https://en.wikipedia.org/wiki/Skewness)

1. A normal distribution and any other symmetric distribution with finite third moment has a skewness of 0
2. A half-normal distribution has a skewness just below 1
3. An exponential distribution has a skewness of 2
4. A lognormal distribution can have a skewness of any positive value, depending on its parameters

Using the above guidelines I'll apply a log transform to any column whose value is greater than or equal to 1.

```python
num.skew()
```

I need to ensure that I'm not log transforming values which are zero (eg: many people will 
have zero as their minimum loyalty discount), I'll also need to exclude the zero values when transforming the 
numbers back to real units.  

In order to do this I'm going to create a list of columns which meet my criteria for a log transform, then create a function that will log transform all values in those columns which are not equal to zero.

```python
# Creating a list of columns to be transformed
change = ['CampaignsRedeemed', 'Percent_CampaignRedeemed', 'CouponRedeemed_Count', 'Percent_CouponsRedeemed',
         'TypeB', 'TypeC', 'PercentTypeC', 'Count_products', '%_purchases_product_mailer', 'Avg_quantity_purchased'
         ,'Avg_basket_sales_value', 'Max_basket_sales_value', 'Min_basket_sales_value', 'Total_num_baskets', 
          'Avg_num_items', 'Max_num_items', 'Min_num_items', 'Avg_manu_disc', 'Max_manu_disc', 'Avg_coupon_disc',
         'Max_coupon_disc', 'Avg_loyalty_disc', 'Min_loyal_disc', 'Max_loyal_disc', 'Num_stores_visited',
         'Recency', 'Frequency', 'Monetary', 'Avg_trips_week', 'Avg_shopping_lag']
```

```python
# Defining a function that will log transform any value that is non-zero
def nlog(x):
    if x == 0: return 0
    else: return np.log(x)
```

```python
for i in change:
    data[i] = num[i].map(nlog)
```
